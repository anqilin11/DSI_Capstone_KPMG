```python
df_meta = df[['gas_key', 'gas_name','gas_lat', 'gas_long','attr_name','attr_lat', 'attr_long','crime_coord','crime_county','exit_name', 'exit_lat', 'exit_long', 'distance_to_nearest_exit','nri_geoid','nri_county','nri_risk_rating','nri_intpt_lat', 'nri_intpt_long', 'nri_zipcode']]
df_train = df.drop(columns=['gas_name','attr_name','attr_lat', 'attr_long','crime_coord','crime_county','exit_name', 'exit_lat', 'exit_long', 'distance_to_nearest_exit','nri_county','nri_risk_rating','nri_intpt_lat', 'nri_intpt_long', 'nri_zipcode'])
```

```python
df_train.set_index('gas_key', inplace=True)
print(df_train.columns.values)
print(df_train.shape)
```

```python
df_train.isna().sum().sort_values(ascending = False)
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots(5,6, figsize=(25, 25))
for (i,col) in enumerate(df_train.columns.values[2:]):
    # print((i//6,i%6))
    sns.histplot(ax=ax[i//6][i%6], data=df_train, x=col, bins=20)
    ax[i//6][i%6].set_title(col)

plt.savefig('plot/histplot_attr_dist.png')
```
