```python
def get_avg_from_result(result):
    filtered = pd.DataFrame(result['searchResults'])
    #print(filtered['name'].astype(int).mean())
    return filtered['key'], filtered['shapePoints'], filtered['name'].astype(int).mean()

def get_max_from_result(result):
    filtered = pd.DataFrame(result['searchResults'])
    #print(filtered['name'].astype(int).mean())
    return filtered['key'], filtered['shapePoints'], filtered['name'].astype(int).max()
```

```python
# for each gas station, compute its traffic 
gas_station_traffic_data = []
for gas_i in range(gas_stations_all.shape[0]):
  gas_data = gas_stations_all.iloc[gas_i]
  gas_origin = [gas_data['Latitude'], gas_data['Longitude']]
  # 1 mile radius traffic count
  result_r1 = post_radius_search_request(gas_origin, getRemoteTrafficDataList, 10)

  if result_r1['resultsCount'] != 0:
    key_1m_avg, shape_1m_avg, traff_cnt_1m_avg = get_avg_from_result(result_r1)
    key_1m_max, shape_1m_max, traff_cnt_1m_max = get_max_from_result(result_r1)
  else:
    traff_cnt_1m_avg = 0
    traff_cnt_1m_max = 0

  # 5 miles radius traffic count
  result_r5 = post_radius_search_request(gas_origin, getRemoteTrafficDataList, 5)
  if result_r5['resultsCount'] != 0:
    key_5m_avg, shape_5m_avg, traff_cnt_5m_avg = get_avg_from_result(result_r5)
    key_5m_max, shape_5m_max, traff_cnt_5m_max = get_max_from_result(result_r5)
  else:
    traff_cnt_5m_avg = 0
    traff_cnt_5m_max = 0

  # print(result_r5)

  
  gas_station_traffic_data.append([gas_i, gas_data['name'], gas_data['Latitude'],gas_data['Longitude'],
                                traff_cnt_1m_avg, traff_cnt_1m_max, traff_cnt_5m_avg, traff_cnt_5m_max])
```

```python
gas_station_traffic_data = pd.DataFrame(gas_station_traffic_data)
gas_station_traffic_data.columns = ['gas_key', 'gas_name', 'gas_lat','gas_long',
                        'traff_cnt_10m_avg', 'traff_cnt_10m_max', 'traff_cnt_5m_avg', 'traff_cnt_5m_max']
```
