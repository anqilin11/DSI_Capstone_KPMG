```python
%cd /content/drive/My Drive/Capstone-KPMG/Data_Exploration/1_ACS
df_income = pd.read_csv("ACSST5Y2020.S1901-Data.csv", header = 1)
df_income.head()
```

```python
df_income_use = df_income[['Geographic Area Name', 'Estimate!!Households!!Median income (dollars)']]
df_income_use.columns = ['zip_income', 'household_median_income']
df_income_use['zip_income'] = df_income_use['zip_income'].str.split(" ",expand=True)[1].astype(int)
df_income_use.head()
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="geoapiExercises")
location = geolocator.reverse(str(48.393760)+","+str(-122.329806)).raw['address'].get('postcode', '')
location
```

```python
df_use[['gas_lat', 'gas_long', 'crime_coord']]
```

```python
df_use['zip'] = df_use['crime_coord'].apply(lambda x: geolocator.reverse(x).raw['address'].get('postcode', ''))
df_use['zip'].head()
```

```python
df_income_use.head()
```

```python
df_income_use["household_median_income"] = df_income_use["household_median_income"].apply(lambda x: np.nan if x == '-' else x)
df_income_use["household_median_income"] = pd.to_numeric(df_income_use["household_median_income"])
df_income_use.head()
```

```python
df_income_use.dtypes
```

```python
df_final = df_use.merge(df_income_use, how = 'left', left_on = 'zip', right_on = 'zip_income')
```

```python
df_final
```

```python
df_final["household_median_income"] = np.where(df_final["household_median_income"] == "-", np.nan, df_use["household_median_income"])
```
